# **3. Boxenplot of Salary Distribution**


# Explanation: This shows the overall distribution of salaries using a boxenplot.


# The code now includes a line to calculate the AverageSalary column by taking the average of the Min Salary and Max Salary columns.

```python
#dataset_df[['Lower Salary(K)', 'Upper Salary(K)', 'Job_Portal']] = dataset_df['Salary Estimate'].str.extract(r'(\d+)K-\$(\d+)K \((.*)\)')
```

#  this code is to find strings like "$30K-$$30K-$50K" and capture the "30K" and "50K" parts.

# this code snippet uses a regular expression to extract the minimum and maximum salary values from the Salary Estimate column and stores them in new columns called Min Salary and Max Salary. This is a common data cleaning step to make the salary information more structured and easier to work with.

```python
# Extract minimum and maximum salary from the 'Salary Estimate' column
salary_pattern = r'\$([\dK]+)-\$([\dK]+)'
dataset_df[['Min Salary', 'Max Salary']] = dataset_df['Salary Estimate'].str.extract(salary_pattern)
```

# this code ensures that the 'Min Salary' and 'Max Salary' columns in the DataFrame contain numerical salary values instead of strings, making them suitable for calculations and analysis.

```python
def convert_salary(salary):
    if 'K' in salary:
        return int(salary.replace('K', '')) * 1000
    return int(salary)

dataset_df['Min Salary'] = dataset_df['Min Salary'].fillna('0').apply(convert_salary)
dataset_df['Max Salary'] = dataset_df['Max Salary'].fillna('0').apply(convert_salary)
```

```python
print(dataset_df.columns)
```

```python
dataset_df['AverageSalary'] = (dataset_df['Min Salary'] + dataset_df['Max Salary']) / 2
```

```python
sns.boxenplot(data=dataset_df, x='AverageSalary')
plt.xlabel('Average Salary')
plt.ylabel('Count')
plt.title('Distribution of Average Salary')
plt.show()
```

```python
sns.histplot(dataset_df['AverageSalary'], bins=20)
plt.title('Overall Salary Distribution')
plt.xlabel('Average Salary')
plt.ylabel('Job Posting Count')  # More descriptive y-axis label

# Customize x-axis ticks
plt.xticks([40000, 60000, 80000, 100000, 120000, 140000],
           ['40k', '60k', '80k', '100k', '120k', '140k'])

# Format y-axis ticks to display as integers
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()
```
